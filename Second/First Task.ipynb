```python
#imports
from Crypto.Cipher import DES, DES3, AES
from Crypto.Util.Padding import pad, unpad
import codecs
```

```python
#Common Functions
def diff_bits(a,b):
    bitsdiff = "UnEqual Bits!" if len(a) != len(b) else sum ( a[i] != b[i] for i in range(len(a))) 
    return bitsdiff,len(a),len(b)

def hex_bin(inhex):
    return ''.join(bin(int(c, 16))[2:].zfill(4) for c in str(inhex)[2:-1])

def encrypt(cipher,msg,padding):
#     ciphertext = cipher.encrypt(bytes.fromhex(msg))
    ciphertext = cipher.encrypt(pad(msg.encode(),padding))
    return ciphertext

def decrypt(cipher,ciphertext,padding):
    plaintext = cipher.decrypt(ciphertext)
#     return binascii.hexlify(plaintext).decode()
    plaintext=unpad(plaintext,padding)
    return plaintext.decode()

def display(ciphertext, ciphertext_hex, ciphertext_bin, plaintext, oldcipher):
    print('Cipher text: ',ciphertext_hex)
    print('Cipher text (Binary): ',ciphertext_bin)
    if not plaintext:
        print('Message is corrupted!')
    else:
        print('Plain text (Decrypted):', plaintext)
    if oldcipher=="":
        oldcipher=ciphertext_bin
    else:
        print("Bits Difference: ",diff_bits(oldcipher,ciphertext_bin))
        oldcipher=ciphertext_bin
```

```python
des_key = b'Moizhere'
des3_key= b'AbdulMoizQuddusx'
des3_key = DES3.adjust_key_parity(des3_key)
aes_key= b'AbdulMoizQuddusx'
# key = bytes.fromhex('0f1571c947d9e859')

def DES_(key,inputtext):
    cipher = DES.new(key, DES.MODE_ECB)
    ciphertext = encrypt(cipher,inputtext,padding=8)
    ciphertext_hex = codecs.encode(ciphertext,'hex') # to hex
    ciphertext_bin =  hex_bin(ciphertext_hex) # to Bin
    cipher = DES.new(key, DES.MODE_ECB)
    plaintext = decrypt(cipher,ciphertext,padding=8) 
    return ciphertext, ciphertext_hex, ciphertext_bin, plaintext

def DES3_(key,inputtext):
    cipher = DES3.new(key, DES3.MODE_ECB)
    ciphertext = encrypt(cipher,inputtext,padding=8)
    ciphertext_hex = codecs.encode(ciphertext,'hex') # to hex
    ciphertext_bin =  hex_bin(ciphertext_hex) # to Bin
    cipher = DES3.new(key, DES3.MODE_ECB)
    plaintext = decrypt(cipher,ciphertext,padding=8) 
    return ciphertext, ciphertext_hex, ciphertext_bin, plaintext    

def AES_(key,inputtext):
    cipher = AES.new(key, AES.MODE_ECB)
    ciphertext = encrypt(cipher,inputtext,padding=16)
    ciphertext_hex = codecs.encode(ciphertext,'hex') # to hex
    ciphertext_bin =  hex_bin(ciphertext_hex) # to Bin
    cipher = AES.new(key, AES.MODE_ECB)
    plaintext = decrypt(cipher,ciphertext,padding=16) 
    return ciphertext, ciphertext_hex, ciphertext_bin, plaintext
```

```python
#RSA
import rsa

def generate_keys():
    (pubKey, privKey) = rsa.newkeys(512)
    with open('keys/pubkey.pem', 'wb') as f:
        f.write(pubKey.save_pkcs1('PEM'))

    with open('keys/privkey.pem', 'wb') as f:
        f.write(privKey.save_pkcs1('PEM'))

def load_keys():
    with open('keys/pubkey.pem', 'rb') as f:
        pubKey = rsa.PublicKey.load_pkcs1(f.read())

    with open('keys/privkey.pem', 'rb') as f:
        privKey = rsa.PrivateKey.load_pkcs1(f.read())

    return pubKey, privKey

def encrypt(msg, key):
    return rsa.encrypt(msg.encode(), key)

def decrypt(ciphertext, key):
    try:
        return rsa.decrypt(ciphertext, key).decode()
    except:
        return False

def sign_sha256(msg, key):
    return rsa.sign(msg.encode(), key, 'SHA-256')

def verify_sha1(msg, signature, key):
    try:
        return rsa.verify(msg.encode(), signature, key) == 'SHA-256'
    except:
        return False

generate_keys()
pubKey, privKey = load_keys()
print(pubKey,privKey)
message = input('Enter a message:')
ciphertext = encrypt(message, pubKey)

signature = sign_sha256(message, privKey)

plaintext = decrypt(ciphertext, privKey)

print(f'Cipher text: {ciphertext}')
print(f'Signature: {signature}')

if plaintext:
    print(f'Plain text: {plaintext}')
else:
    print('Could not decrypt the message.')

if verify_sha1(plaintext, signature, pubKey):
    print('Signature verified!')
else:
    print('Could not verify the message signature.')
```
